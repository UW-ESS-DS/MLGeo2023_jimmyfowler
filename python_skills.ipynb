```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make array for x axis
x = np.linspace(0, 2*np.pi, 100)

# lambda function sin(x) for y axis
f1 = lambda x: np.sin(x)

# basic function cos(x)
def f2(x):
    return np.cos(x)

# another lambda function, tan(x)
f3 = lambda x: 0.02*np.tan(x)

# plot x, sin(x), cos(x) and tan(x)
plt.plot(x, f1(x), 'm--', label='sin(x)')
plt.plot(x, f2(x), 'b-', label='cos(x)')
plt.plot(x, f3(x), 'r:', label='0.02tan(x)')

# add grid, title, legend and axes labels
plt.grid()
plt.title('A Trig Function Display')
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.legend(loc='upper center')

# save the figure
plt.savefig('trig_functions')
```
